## Data Cleaning

### Filtering aircrafts and events

We want to filter the dataset to include aircraft that the client is interested in an analysis of:
- inspect relevant columns
- figure out any reasonable imputations
- filter the dataset

```python
df.columns.tolist
```

```python
#filter for last 40 years to represent active makes
df['Event.Date']=pd.to_datetime(df['Event.Date'])
df=df[df['Event.Date']>='1983-01-01']
```

```python
#filter for profesional builds only
df=df[df['Amateur.Built']== 'No']
```

```python
#Filter for Category 'Airplane' or records where category is missing (likely small planes)
df = df[(df['Aircraft.Category'] == 'Airplane') | (df['Aircraft.Category'].isna())]

# Clean Make strings to help with filtering
df['Make'] = df['Make'].str.upper().str.strip()

# Explicitly remove known helicopter manufacturers that might be in the NaN category
heli_makes = ['BELL', 'HUGHES', 'ROBINSON', 'SIKORSKY', 'EUROCOPTER', 'AEROSPATIALE', 
              'ENSTROM', 'SCHWEIZER', 'MD HELICOPTERS', 'AGUSTA', 'HILLER']
df = df[~df['Make'].isin(heli_makes)]
```

### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
#Impute injury NaNs with 0 (assuming NaN implies no report of such injury)
injury_cols = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']
for col in injury_cols:
    df[col] = df[col].fillna(0)

# Calculate Total Occupants
df['Total.Occupants'] = df[injury_cols].sum(axis=1)

# Create Severe Injury metrics
df['Severe.Injury.Count'] = df['Total.Fatal.Injuries'] + df['Total.Serious.Injuries']

# Likelihood of severe injury (Severe Injuries / Total Occupants)
# We use np.where to handle cases where Total.Occupants is 0
df['Severe.Injury.Rate'] = np.where(df['Total.Occupants'] > 0, 
                                    df['Severe.Injury.Count'] / df['Total.Occupants'], 
                                    0)
```

**Aircraft.Damage**
- identify and execute any cleaning tasks
- construct a derived column tracking whether an aircraft was destroyed or not.

```python
#Handle missing damage values
df['Aircraft.damage'] = df['Aircraft.damage'].fillna('Unknown')

# Create binary column: 1 if Destroyed, 0 otherwise
df['is_destroyed'] = (df['Aircraft.damage'] == 'Destroyed').astype(int)
```

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

```python
#Filter for robustness: Keep Makes with at least 50 recorded accidents
make_counts = df['Make'].value_counts()
valid_makes = make_counts[make_counts >= 50].index
df = df[df['Make'].isin(valid_makes)]

print(f"Number of robust Makes remaining: {len(valid_makes)}")
```

### Inspect Model column
- Get rid of any NaNs.
- Inspect the column and counts for each model/make. Are model labels unique to each make?
- If not, create a derived column that is a unique identifier for a given plane type.

```python
# Standardize Model strings and handle NaNs
df['Model'] = df['Model'].fillna('Unknown').str.upper().str.strip()

# Create a unique identifier for Make-Model pairs
df['Make_Model'] = df['Make'] + " " + df['Model']

# Inspect counts for unique plane types
print(df['Make_Model'].value_counts().head(10))
```

### Cleaning other columns
- there are other columns containing data that might be related to the outcome of an accident. We list a few here:
- Engine.Type
- Weather.Condition
- Number.of.Engines
- Purpose.of.flight
- Broad.phase.of.flight

Inspect and identify potential cleaning tasks in each of the above columns. Execute those cleaning tasks. 

**Note**: You do not necessarily need to impute or drop NaNs here.

```python
df['Weather.Condition'] = df['Weather.Condition'].fillna('UNK').str.upper()
df['Broad.phase.of.flight'] = df['Broad.phase.of.flight'].fillna('UNKNOWN').str.upper()
df['Engine.Type'] = df['Engine.Type'].fillna('UNKNOWN')
df['Purpose.of.flight'] = df['Purpose.of.flight'].fillna('Unknown')
```

### Column Removal
- inspect the dataframe and drop any columns that have too many NaNs

```python
# Drop columns with more than 50% missing data
nan_pct = df.isnull().mean()
cols_to_drop = nan_pct[nan_pct > 0.5].index.tolist()

# Add redundant or high-cardinality ID columns that won't benefit statistical modeling
cols_to_drop.extend(['Event.Id', 'Accident.Number', 'Registration.Number', 'Publication.Date'])

# Execute removal
df.drop(columns=[c for c in cols_to_drop if c in df.columns], inplace=True)
```
